```python
#subplots()
#Multiple plots in one figure
#1st option:Create subplot()

fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                      ncols=2,
                                      figsize=(10,15))
#Plot data to each axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(butter_price.keys(),butter_price.values());
ax4.hist(np.random.randn(110000));
```

```python
#Create Options2:
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#Index to plot the data
ax[0,0].plot(x,x/2);
ax[0,1].bar(butter_price.keys(),butter_price.values());
ax[1,0].scatter(np.random.random(10),np.random.random(10));
ax[1,1].hist(np.random.randn(110000));
```

```python
over_45=heart_disease[heart_disease["age"]>45]
over_45
```

```python
over_45.plot(kind="scatter",x="age",y="chol",c="target",figsize=(10,6));
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_45.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
ax.set_xlim([45,100])
```

```python
#Making a bit  more of a complicated Plot

fig,ax=plt.subplots(figsize=(10,6))


#plot the Data
scatter=ax.scatter(over_45["age"],
                   over_45["chol"],
                   c=over_45["target"])

#customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol");

ax.legend(*scatter.legend_elements(),title="Target");      




```

```python
#Adding another plot to existing styled one
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           sharex=True,
                           figsize=(10,8))

#add data for Ax0
scatter=ax0.scatter(over_45["age"],
                   over_45["chol"],
                   c=over_45["target"])

#Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
  
       ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(),title="Target"); 

#Setup a mean line

ax0.axhline(y=over_45["chol"].mean(),
            color="r",
            linestyle="--",
            label="Average")
#Add data for ax1

scatter=ax1.scatter(over_45["age"],
                   over_45["thalach"],
                   c=over_45["target"])

#Customize ax1
ax1.set(title="Heart Disease and Cholestrol Levels",
  
       xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(),title="Target");

#Setup a mean line

ax1.axhline(y=over_45["thalach"].mean(),
            color="b",
            linestyle="--",
            label="Average")

#Title
fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight="bold");



```

```python
#Adding another plot to existing styled one
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           sharex=True,
                           figsize=(10,8))

#add data for Ax0
scatter=ax0.scatter(over_45["age"],
                   over_45["chol"],
                   c=over_45["target"])

#Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
  
       ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(),title="Target"); 

#Setup a mean line

ax0.axhline(y=over_45["chol"].mean(),
            color="r",
            linestyle="--",
            label="Average")
#Add data for ax1

scatter=ax1.scatter(over_45["age"],
                   over_45["thalach"],
                   c=over_45["target"])

#Customize ax1
ax1.set(title="Heart Disease and Cholestrol Levels",
  
       xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(),title="Target");

#Setup a mean line

ax1.axhline(y=over_45["thalach"].mean(),
            color="b",
            linestyle="--",
            label="Average")

#Title
fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight="bold");
```

```python
#After Changing
#Adding another plot to existing styled one
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           sharex=True,
                           figsize=(10,8))

#add data for Ax0
scatter=ax0.scatter(over_45["age"],
                   over_45["chol"],
                   c=over_45["target"],
                    cmap='winter')

#Customize ax0
#Set x axis
ax0.set_xlim([50,80])
ax0.set(title="Heart Disease and Cholestrol Levels",
  
       ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(),title="Target"); 

#Setup a mean line

ax0.axhline(y=over_45["chol"].mean(),
            color="r",
            linestyle="--",
            label="Average")
#Add data for ax1

scatter=ax1.scatter(over_45["age"],
                   over_45["thalach"],
                   c=over_45["target"])

#Customize ax1
ax1.set(title="Heart Disease and Cholestrol Levels",
  
       xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(),title="Target");

#set the y-axis

ax1.set_ylim([60,200])

#Setup a mean line

ax1.axhline(y=over_45["thalach"].mean(),
            color="b",
            linestyle="--",
            label="Average")

#Title
fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight="bold");
```
